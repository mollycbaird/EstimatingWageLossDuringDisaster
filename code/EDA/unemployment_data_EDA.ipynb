```python
import pandas as pd
```

```python
# read in the data from the labor statistics site
# https://www.bls.gov/web/laus/ststdsadata.txt
# originally from this site: https://www.bls.gov/lau/rdscnp16.htm

df = pd.read_excel('../data/ststdsadata.xlsx')
```

```python
df.head(20)
```

```python
# this was saved in a weird format, 
# the columns weren't correctly detected
# renaming here 
# civ_lf means "civilian labor force"

col_dict = {'States and selected areas:  Employment status of the civilian noninstitutional population,':'FIPS_Code',
            'Unnamed: 1':'state_and_area',
            'Unnamed: 2':'period_year',
            'Unnamed: 3':'period_month',
            'Unnamed: 4':'area_population',
            'Unnamed: 5':'civ_lf_total',
            'Unnamed: 6':'civ_lf_total_%',
            'Unnamed: 7':'civ_lf_employed',
            'Unnamed: 8':'civ_lf_employed_%',
            'Unnamed: 9':'civ_lf_unemployed',
            'Unnamed: 10':'civ_lf_unemployed_%'
           }
df_renamed = df.rename(columns = col_dict)
```

```python
df_renamed.head(10)
```

```python
# dropping indicies 0 through 6

df_renamed_dropped = df_renamed.drop(index = list(range(7)))
```

```python
# making sure it worked 

df_renamed_dropped.head()
```

```python
# exporting to the data folder

df_renamed_dropped.to_csv('../data/unemployment_by_state_1976.csv', index = False)
```
